```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def in_degree_0(G):
	S = []
	nodes = [node for (node, val) in G.nodes().items()]
	for n in nodes:
		if G.in_degree(n) == 0:
			S.append(n)
	return S
```

```python
def topological_sort(G):
	L = []
	S = in_degree_0(G)
	while S != []:
		node = S[0]
		L.append(node)
		G.remove_node(node)
		S = in_degree_0(G)
	if G.number_of_edges() > 0:
		return 'G contiene ciclos!'
	return 'Orden: ' + ' '.join(str(e) for e in L)
```

```python
G = nx.DiGraph() 
G.add_nodes_from(list(range(1,8)))
G.add_edges_from([(1,2),(1,3),(1,4),(2,3),(3,5),(3,6),(4,6),(5,7),(6,5),(6,7)])
```

```python
nx.draw(G, with_labels = True)
plt.show()
```

```python
topological_sort(G)
```
