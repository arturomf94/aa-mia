```python
import networkx as nx
import matplotlib.pyplot as plt
import sys
```

```python
fig_size = plt.rcParams["figure.figsize"]
plt.rcParams["figure.figsize"] = [10,8]
```

```python
N = 50
G = nx.grid_2d_graph(N,N)
pos = dict( (n, n) for n in G.nodes() )
labels = dict( ((i, j), (i, j)) for i, j in G.nodes() )
new_edges =[]
for i, j in G.nodes():
    if(i < N - 1 and j < N - 1):
        new_edges.append(((i,j),(i + 1, j + 1)))
    if(i < N - 1 and j > 0):
        new_edges.append(((i,j),(i + 1, j - 1)))
G.add_edges_from(new_edges)
```

```python
# Draw AI in Grid

for i in range(8):
    G.remove_node((5 + i, 20 + i))
    G.remove_node((6 + i, 20 + i))
    G.remove_node((7 + i, 20 + i))
    
for i in range(8):
    G.remove_node((15 + i,27 - i))
    G.remove_node((16 + i,27 - i))
    G.remove_node((17 + i,27 - i))
    
for i in range(10):
    G.remove_node((10 + i,22))
    
for i in range(8):
    G.remove_node((36, 20 + i))
    G.remove_node((37, 20 + i))
    G.remove_node((38, 20 + i))
    
for i in range(11):
    G.remove_node((32 + i,19))
    
for i in range(11):
    G.remove_node((32 + i,28))
```

```python
nx.draw(G, pos, node_size = 10)
plt.show()
```

```python
def get_connected_component(G, node):
    G_new = G.copy()
    all_components = list(nx.connected_components(G))
    for component in all_components:
        if node in component:
            nonconnected = [element for element in all_components if element != component]
            for other_component in nonconnected: 
                G_new.remove_nodes_from(other_component)
            return G_new
    return nx.Graph()
G_new = get_connected_component(G, (0,0))
nx.draw(G_new, pos, node_size = 10)
plt.show()
```

```python
def dijkstra(G, source, goal):
    G_working = get_connected_component(G, source)
    nodes = [node for node in G_working.nodes()]
    if not goal in nodes:
        return []
    distances = [sys.maxint]*len(nodes)
    previous = [(sys.maxint, sys.maxint)]*len(nodes)
    Q = [node for node in G_working.nodes()]
    distances[nodes.index(source)] = 0
    while Q != []:
        indices = [nodes.index(elem) for elem in Q]
        filtered_distances = [distances[i] for i in indices]
        min_dist = min(filtered_distances)
        indices_with_min_dist = [i for i,x in enumerate(distances) if x == min_dist]
        for index_with_min_dist in indices_with_min_dist:
            u = nodes[index_with_min_dist]
            if u in Q:
                Q.remove(u)
                break
        if u == goal:
            break
        for neighbor in list(G_working.neighbors(u)): 
            if neighbor in Q:
                neighbor_index = nodes.index(neighbor)
                alt = min_dist + 1
                if alt < distances[neighbor_index]:
                    distances[neighbor_index] = alt
                    previous[neighbor_index] = u
    
    S = []
    u = goal
    goal_index = nodes.index(u)
    if previous[goal_index] != (sys.maxint, sys.maxint) or u == source:
        while u != (sys.maxint, sys.maxint):
            goal_index = nodes.index(u)
            S.insert(0,u)
            u = previous[goal_index]
    
    return S
```

```python
path_to_goal = dijkstra(G,(0,0),(49,49))
nx.draw(G, pos, node_color='r', node_size=10)
nx.draw(G, pos,nodelist=path_to_goal,node_color='b',node_size=50)
```
