```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
fig_size = plt.rcParams["figure.figsize"]
plt.rcParams["figure.figsize"] = [10,8]
```

```python
N = 50
G = nx.grid_2d_graph(N,N)
pos = dict( (n, n) for n in G.nodes() )
labels = dict( ((i, j), (i, j)) for i, j in G.nodes() )
new_edges =[]
for i, j in G.nodes():
    if(i < N - 1 and j < N - 1):
        new_edges.append(((i,j),(i + 1, j + 1)))
    if(i < N - 1 and j > 0):
        new_edges.append(((i,j),(i + 1, j - 1)))
G.add_edges_from(new_edges)
```

```python
# Draw AI in Grid

for i in range(8):
    G.remove_node((5 + i, 20 + i))
    G.remove_node((6 + i, 20 + i))
    G.remove_node((7 + i, 20 + i))
    
for i in range(8):
    G.remove_node((15 + i,27 - i))
    G.remove_node((16 + i,27 - i))
    G.remove_node((17 + i,27 - i))
    
for i in range(10):
    G.remove_node((10 + i,22))
    
for i in range(8):
    G.remove_node((36, 20 + i))
    G.remove_node((37, 20 + i))
    G.remove_node((38, 20 + i))
    
for i in range(11):
    G.remove_node((32 + i,19))
    
for i in range(11):
    G.remove_node((32 + i,28))
```

```python
nx.draw(G, pos, node_size = 10)
plt.show()
```
