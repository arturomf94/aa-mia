```python
# Primes!
x = np.arange(1,5001)
y5 = x*np.log(x)

count = 0
```

```python
def prime(n):
    global count

    if n <= 1:
        return False
    count += 1
    if n == 2:
        return True 
    count += 1  
    if n % 2 == 0:
        return False
    count += 1
    d = 3
    while d * d <= n:
        if (n % d == 0):
            return False
        count += 1
        d += 2 
    return True
```

```python
def count_primes_division(n):
    global count
    primes = 0
    i = 1
    while(i <= n):
        count += 1
        if prime(i):
            primes += 1
        i += 1
    return primes
```

```python
steps_primes_division = []

for n in x:
	count = 0
	count_primes_division(n)
	steps_primes_division.append(count)
```

```python
fig_functions = plt.figure()
ax1 = fig_functions.add_subplot(111)
ax1.plot(x, y5, c = 'm', label = 'n*ln(n)') 
ax1.plot(x, steps_primes_division, c = 'y', label = 'Primos con Division') 
plt.legend(loc='upper left')
plt.title('Comparacion de Funciones')
plt.xlabel('N')
plt.ylabel('Y')
plt.show()
```

```python
def primes_division_complexity(n):
	sum = 0
	for i in range(1,n+1):
		sum = sum + math.sqrt(i) / 2
	return sum

y7 = [primes_division_complexity(n) for n in x]
```

```python
fig_functions = plt.figure()
ax1 = fig_functions.add_subplot(111)
ax1.plot(x, y5, c = 'm', label = 'n*ln(n)') 
ax1.plot(x, steps_primes_division, c = 'y', label = 'Primos con Division')
ax1.plot(x, y7, c = 'r', label = 'Primos con Division: Cota') 
plt.legend(loc='upper left')
plt.title('Comparacion de Funciones')
plt.xlabel('N')
plt.ylabel('Y')
plt.show()
```
